```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_validate
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# 1. Load the dataset directly from a reliable public URL (saves you from uploading a CSV manually)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/breast-cancer-wisconsin.data"
column_names = ['id', 'clump_thickness', 'uniform_cell_size', 'uniform_cell_shape',
                'marginal_adhesion', 'single_epithelial_size', 'bare_nuclei',
                'bland_chromatin', 'normal_nucleoli', 'mitoses', 'class']

df = pd.read_csv(url, names=column_names)

# 2. Clean the data (The 'bare_nuclei' column has some '?' missing values)
df = df.replace('?', np.nan)
df = df.dropna()

# Drop the 'id' column as it's not a feature, and separate features (X) from labels (y)
X = df.drop(columns=['id', 'class'])
y = df['class'] # 2 for benign, 4 for malignant

# 3. Initialize the models
dt_model = DecisionTreeClassifier(
    criterion='entropy',       # J48/C4.5 uses Info Gain / Entropy, not Gini
    min_samples_leaf=2,        # Weka default minNumObj = 2
    random_state=42
)

rf_model = RandomForestClassifier(
    n_estimators=10,           # The 2012 Weka default for Random Forest was 10 trees
    criterion='entropy',
    random_state=42
)

# 4. Perform 10-Fold Cross-Validation collecting Accuracy, Precision, and Recall
scoring_metrics = ['accuracy', 'precision_macro', 'recall_macro', 'f1_macro']

dt_results = cross_validate(dt_model, X, y, cv=10, scoring=scoring_metrics)
rf_results = cross_validate(rf_model, X, y, cv=10, scoring=scoring_metrics)

# 5. Print out the results formatted cleanly
print("--- DECISION TREE RESULTS ---")
print(f"Accuracy:  {dt_results['test_accuracy'].mean() * 100:.2f}%  (Paper J48: 94.56%)")
print(f"Precision: {dt_results['test_precision_macro'].mean():.3f}    (Paper J48: 0.946)")
print(f"Recall:    {dt_results['test_recall_macro'].mean():.3f}    (Paper J48: 0.946)")
print(f"F-Measure: {dt_results['test_f1_macro'].mean():.3f}    (Paper J48: 0.946)\n")

print("--- RANDOM FOREST RESULTS ---")
print(f"Accuracy:  {rf_results['test_accuracy'].mean() * 100:.2f}%  (Paper RF: 96.13%)")
print(f"Precision: {rf_results['test_precision_macro'].mean():.3f}    (Paper RF: 0.962)")
print(f"Recall:    {rf_results['test_recall_macro'].mean():.3f}    (Paper RF: 0.961)")
print(f"F-Measure: {rf_results['test_f1_macro'].mean():.3f}    (Paper RF: 0.961)")
```
